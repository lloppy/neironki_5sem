```python
# Определим числовые столбцы
numeric_cols = X.select_dtypes(include=['int64', 'float64']).columns.tolist()
print("Числовые столбцы:", numeric_cols)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

# Применяем только к числовым признакам
X[numeric_cols] = scaler.fit_transform(X[numeric_cols])
```

Масштабирование должно выполняться после разделения данных на обучающую и тестовую выборки, чтобы избежать утечки данных. Здесь мы показываем общий пример, но в дальнейшем следует перенести масштабирование в pipeline.

Отбор признаков и снижение размерности

SelectKBest: Выбирает K лучших признаков по определённой метрике.

```python
from sklearn.feature_selection import SelectKBest, f_classif

# Выбираем 1000 лучших признаков по ANOVA F-тесту
selector = SelectKBest(score_func=f_classif, k=1000)
X_best = selector.fit_transform(X, y)

# Получаем список выбранных признаков
selected_features = X.columns[selector.get_support()]
print(f"Выбранные признаки: {len(selected_features)}")
```

Разделение данных на обучающую и тестовую выборки

```python
y_encoded = y.map({'yes': 1, 'no': 0})
```

```python
from sklearn.model_selection import train_test_split

# Если вы уже выбрали признаки и подготовили X
# Например, если использовали SelectKBest
X_selected = X[selected_features]

# Разделение данных
X_train, X_test, y_train, y_test = train_test_split(
    X_selected, y_encoded, test_size=0.2, random_state=42, stratify=y
)

print(f"Обучающая выборка: {X_train.shape}")
print(f"Тестовая выборка: {X_test.shape}")
print("\nРаспределение классов в y:", y_encoded.value_counts(normalize=True))
```

```python
results = []
plt.figure(figsize=(10, 8))

for name, model in models.items():
    y_pred = model.predict(X_test)
    y_prob = model.predict_proba(X_test)[:,1]  # вероятности для ROC

    # Метрики
    acc = accuracy_score(y_test, y_pred)
    prec = precision_score(y_test, y_pred)
    rec = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    auc = roc_auc_score(y_test, y_prob)

    results.append(pd.DataFrame({
        'Model': [name],
        'Accuracy': [acc],
        'Precision': [prec],
        'Recall': [rec],
        'F1-Score': [f1],
        'ROC-AUC': [auc]
    }))

    # Для ROC-кривой
    fpr, tpr, _ = roc_curve(y_test, y_prob)
    plt.plot(fpr, tpr, label=f"{name} (AUC = {auc:.2f})")

plt.plot([0,1],[0,1], 'k--')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC-кривые для разных моделей")
plt.legend(loc='lower right')
plt.grid(alpha=0.3)
plt.show()

results_df = pd.concat(results, ignore_index=True)
print("\nРезультаты моделей:")
display(results_df)
```

```python
# =====================================
# 1. Импорт библиотек
# =====================================
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

# ML и вспомогательные инструменты
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import (accuracy_score, precision_score, recall_score,
                             f1_score, roc_auc_score, roc_curve, classification_report,
                             confusion_matrix, ConfusionMatrixDisplay)

# Импорт других моделей
from xgboost import XGBClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neural_network import MLPClassifier

# Библиотека для борьбы с дисбалансом
!pip install imblearn
from imblearn.over_sampling import SMOTE
```

```python
# =====================================
# 10. Матрица путаницы и отчёт (Confusion Matrix)
# =====================================
for name, model in models.items():
    print(f"\n=== {name} ===")
    y_pred = model.predict(X_test)

    print(classification_report(y_test, y_pred))

    cm = confusion_matrix(y_test, y_pred)
    disp = ConfusionMatrixDisplay(cm)
    disp.plot(cmap='Blues')
    plt.title(f"Confusion Matrix: {name}")
    plt.show()
```

```python
# =====================================
# 11. Подбор гиперпараметров (пример для RandomForest в Pipeline)
# =====================================
# Пример: создадим Pipeline, где заранее прописан SelectKBest -> RandomForest
# (Если вы уже делали это, пропустите или адаптируйте.)

from sklearn.pipeline import Pipeline

pipeline = Pipeline([
    ('select', SelectKBest(f_classif, k=1000)),  # Например, 1000 лучших генов
    ('clf', RandomForestClassifier(random_state=42))
])

param_grid = {
    'select__k': [500, 1000, 1500],
    'clf__n_estimators': [100, 200],
    'clf__max_depth': [None, 10, 20],
    'clf__min_samples_split': [2, 5],
    'clf__min_samples_leaf': [1, 2]
}

grid_search = GridSearchCV(pipeline, param_grid,
                           cv=5, scoring='roc_auc',
                           n_jobs=-1, verbose=1)

# Важно: для корректного SMOTE в cross-validation, нужно использовать
# imblearn-пайплайн или StratifiedKFold отдельно.
# Для простоты — используем X_train_balanced (уже сбалансированный).
grid_search.fit(X_train_balanced, y_train_balanced)
print("Лучшие параметры:", grid_search.best_params_)
print("Лучший ROC-AUC (CV):", grid_search.best_score_)

best_model = grid_search.best_estimator_

# Проверим на тесте
y_pred_best = best_model.predict(X_test)
y_prob_best = best_model.predict_proba(X_test)[:,1]

print("\nОценка на тесте (лучшая модель RandomForest + SelectKBest):")
print(classification_report(y_test, y_pred_best))
print("ROC-AUC:", roc_auc_score(y_test, y_prob_best))
```

```python
# Создание пайплайна с масштабированием, выборкой признаков и классификатором MLP
pipeline = Pipeline([
    ('scale', StandardScaler()),  # Масштабирование признаков
    ('select', SelectKBest(f_classif, k=1000)),  # Например, 1000 лучших генов
    ('clf', MLPClassifier(max_iter=1000, random_state=42))
])

# Определение сетки гиперпараметров для MLPClassifier
param_grid = {
    'select__k': [500, 1000, 1500],
    'clf__hidden_layer_sizes': [(100,), (100, 50), (50, 50, 50)],
    'clf__activation': ['relu', 'tanh', 'logistic'],
    'clf__solver': ['adam', 'sgd'],
    'clf__alpha': [0.0001, 0.001, 0.01],
    'clf__learning_rate': ['constant', 'adaptive'],
    'clf__learning_rate_init': [0.001, 0.01, 0.1]
}

# Настройка GridSearchCV
grid_search = GridSearchCV(
    estimator=pipeline,
    param_grid=param_grid,
    cv=5,
    scoring='roc_auc',
    n_jobs=-1,
    verbose=2  # Можно увеличить для более подробного вывода
)

# Важно: для корректного применения SMOTE в cross-validation рекомендуется использовать
# пайплайн из imblearn или StratifiedKFold. Для простоты используем X_train_balanced (уже сбалансированный).

# Обучение GridSearchCV
grid_search.fit(X_train_balanced, y_train_balanced)

# Вывод лучших параметров и лучшего значения ROC-AUC
print("Лучшие параметры:", grid_search.best_params_)
print("Лучший ROC-AUC (CV):", grid_search.best_score_)

# Извлечение лучшей модели
best_model = grid_search.best_estimator_

# Оценка на тестовом наборе
y_pred_best = best_model.predict(X_test)
y_prob_best = best_model.predict_proba(X_test)[:, 1]

print("\nОценка на тесте (лучшая модель MLP + SelectKBest):")
print(classification_report(y_test, y_pred_best))
print("ROC-AUC:", roc_auc_score(y_test, y_prob_best))
```

```python
# =====================================
# Визуализация ROC-кривой для лучшей модели
# =====================================

from sklearn.metrics import roc_curve, auc

# Вычисляем FPR, TPR и пороги
fpr, tpr, thresholds = roc_curve(y_test, y_prob_best)
roc_auc = auc(fpr, tpr)

# Построение ROC-кривой
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='darkorange',
         lw=2, label=f'ROC curve (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--', label='Random Classifier (AUC = 0.50)')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate (FPR)')
plt.ylabel('True Positive Rate (TPR)')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc='lower right')
plt.grid(alpha=0.3)
plt.show()
```

```python
# =====================================
# 12. Важность признаков (Permutation Importance) - пример для MLPClassifier
# =====================================
from sklearn.inspection import permutation_importance
from sklearn.neural_network import MLPClassifier

# Если в пайплайне последняя стадия - MLPClassifier (или другая модель без feature_importances_),
# используем Permutation Importance.

if isinstance(best_model.named_steps['clf'], MLPClassifier):
    # Важно: Permutation Importance вычисляют на отдельном наборе (например, тестовом).
    # Здесь X_test, y_test — это уже не балансированные данные, а реальные для итоговой оценки.
    # Если тестовая выборка маленькая, результат может быть нестабильным, можно увеличить n_repeats.
    r = permutation_importance(
        best_model, X_test, y_test,
        n_repeats=10,         # количество повторов перестановки
        random_state=42,
        scoring='roc_auc'     # метрика, по которой считаем важность
    )

    importances = r.importances_mean  # средняя важность по всем перестановкам

    # Получаем список признаков, которые отобрал SelectKBest
    mask = best_model.named_steps['select'].get_support()
    selected_cols = X.columns[mask]

    # Сортируем индексы по убыванию важности
    indices = np.argsort(importances)[::-1]

    print("\nТоп-10 важных признаков (MLP + SelectKBest, Permutation Importance):")
    top_n = 10
    for i in range(top_n):
        idx = indices[i]
        print(f"{i+1}) {selected_cols[idx]}: {importances[idx]:.5f}")

else:
    # Если вдруг попалась модель, у которой есть feature_importances_ (например, RandomForest),
    # используем традиционное извлечение.
    if hasattr(best_model.named_steps['clf'], 'feature_importances_'):
        importances = best_model.named_steps['clf'].feature_importances_
        mask = best_model.named_steps['select'].get_support()
        selected_cols = X.columns[mask]
        indices = np.argsort(importances)[::-1]

        print("\nТоп-10 важных признаков (RandomForest + SelectKBest):")
        for i in range(10):
            idx = indices[i]
            print(f"{i+1}) {selected_cols[idx]}: {importances[idx]:.4f}")
    else:
        print("Важность признаков не доступна для выбранной модели.")
```
